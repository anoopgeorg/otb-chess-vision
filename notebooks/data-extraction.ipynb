```python
base_path = Path().absolute().parent
board_img_path  = base_path / "data/raw/corner-detection/train/images"
```

```python
def load_images(path:Path,count=51):
    print(f"Looking for images in {str(path)}")
    images=[]
    grays=[]
    for i,img_path in enumerate(path.glob('*')):
        if i == (count - 1):
            break
        else :
            img = cv2.imread(str(img_path))
            gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
            images.append(img)
            grays.append(gray)
    print("Image loading done")
    return images,grays
```

```python
imgs,grays= load_images(board_img_path)
```

```python
def resize_image(imgs,size=(416,416)):
    resized_imgs=[]
    for img in imgs:
        if img.shape[:-1] != size:
            rezised_img = cv2.resize(img,size)
            resized_imgs.append(rezised_img)
    return resized_imgs
```

```python
new_imgs = resize_image(imgs)
```

```python
def save_img(imgs):
    for i,img in enumerate(imgs):
        cv2.imwrite(f"{i}.jpg",img)
        
```

```python
save_img(new_imgs)
```
